```python
import requests
i = chr
content_parse =[]
for i in range(ord('A'), ord('Z') + 1):
    temp = requests.get("https://hasset.ukdataservice.ac.uk/hasset/en/index/"+str(chr(i))+"?clang=de")
    content_parse.append(temp.content)
```

```python
complete_content=[]
for i in range(len(content_parse)): 
    complete_content.append([])
    soup = BeautifulSoup(content_parse[i], 'html.parser')
    for li_tag in soup.find_all('ul', {'class':'list-group alphabetical-search-results side-navi'}):
        for span_tag in li_tag.find_all('a'):
            if span_tag.find('span') is None:
                complete_content[i].append((span_tag.text))
            else:
                for span_tag_full in span_tag.find_all('span', {'class':'sr-only'}):
                    complete_content[i].append(span_tag_full.text)
```
